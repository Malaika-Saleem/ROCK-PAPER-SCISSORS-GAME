```python
#task 4: Rock Paper Scissors game
import random
comp_choices=['rock','paper','scissors']
score_user=0
score_comp=0
contd=True
print('WELCOME TO ROCK PAPER SCISSORS GAME!!!')
while contd:
    user=input('\nChoose rock, paper or scissors: ').lower()
    if user not in ["rock", "paper", "scissors"]:
        print("\nInvalid choice. Please choose rock, paper, or scissors.")
    comp=comp_choices[random.randint(0,2)]
    
    if user=='rock':
        if(comp=='rock'):
            print('\nIts a tie!! You chose rock and computer also chose rock')
        elif(comp=='paper'):
            print('\nYou lost!! you chose rock and comp chose paper')
            score_comp+=1
        elif(comp=='scissors'):
            print('\nYou win!! you chose rock and comp chose scissors')
            score_user+=1
            
    elif user=='paper':
        if(comp=='paper'):
            print('\nIts a tie!! You chose paper and computer also chose paper')
        elif(comp=='scissors'):
            print('\nYou lost!! you chose paper and comp chose scissors')
            score_comp+=1
        elif(comp=='rock'):
            print('\nYou win!! you chose paper and comp chose rock')
            score_user+=1
            
    elif user=='scissors':
        if(comp=='scissors'):
            print('\nIts a tie!! You chose scissors and computer also chose scissors')
        elif(comp=='rock'):
            print('\nYou lost!! you chose scissors and comp chose rock')
            score_comp+=1
        elif(comp=='paper'):
            print('\nYou win!! you chose scissors and comp chose paper')
            score_user+=1    
    
    print('\nYour score: ',score_user,"Computer score: ",score_comp)
    choice=input("\nDo you want to continue playing the game(yes/no)").lower()
    if choice=='yes':
        contd=True
    elif choice=='no':
        contd=False
        
print('\nThankyou for playing your final scores are: ')  

if(score_user>score_comp):
    print("\nYou won overall with : ")
    print('\nYour score: ',score_user,"Computer score: ",score_comp)
    
if(score_user<score_comp):
    print("\nYou lost overall with : ")
    print('\nYour score: ',score_user,"Computer score: ",score_comp)

if(score_user==score_comp):
    print("\nYou tied overall with : ")
    print('\nYour score: ',score_user,"Computer score: ",score_comp)
```
